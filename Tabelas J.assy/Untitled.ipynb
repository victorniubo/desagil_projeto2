```python
indice1 = serie[:46]
```

```python
novo = []
for e in indice1:
    if e != "-":
        novo.append(e[2:])
    else:
        novo.append(e)
```

```python
len(novo)
```

```python
indice2 = serie[46:-1]
```

```python
novo2 = []
for e in indice2:
    if e != "-":
        novo2.append(e[3:])
    else:
        novo2.append(e)
```

```python
len(novo2)+len(novo)
```

```python
geral = novo + novo2
teste = []
for e in geral:
    if e !="-":
        teste.append(e.replace("-","_"))
    else:
        teste.append(e)
```

```python
list = pd.Series(teste, name="x")
list = list.str.lower()
```

```python
list
```

```python
x = df2.join(list)
```
